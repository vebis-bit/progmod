## lesing og plotting av data

vi kan lese filer ved funksjonen _loadtext_:

data = loadtxt('filnavn', skiprows=1, delimiter',', dtype=str, usecols=(1,2,4))

mean() viser gjennomsnitt, std() viser standard avvik.

programmer som bruker .txt filer blir laget sånn


```python
from pylab import *

data = loadtxt('testdata.txt',skiprows=1,delimiter=',')

nr = data[:,0]
hoyde = data[:,1]

plot(nr, hoyde)
show()

snitt = mean(hoyde)
avvik = std(hoyde)

print('snitt', snitt, 'avvik', avvik)
```

```python
from pylab import *

data = loadtxt('solflekkertest.txt',skiprows=1,delimiter=',')

maaneder = data[:,0]
solflekker = data[:,1]

plot(maaneder, solflekker)
grid()
xlabel('år')
ylabel('antall solflekker')
show()

snitt = mean(solflekker)
avvik = std(solflekker)

print('snitt', snitt, 'avvik', avvik)
```

## Regresjon

betydning : Regresjon betyr å finne en funksjon som er best mulig tilpasset alle punktene i et datasett.

problem: for få datapunkter

regresjon av Natriumclorid

```python
from pylab import * 
T = [0, 20, 40, 60, 80, 100]
sol_NH3 = [88.5, 56.0, 34.0, 20.0, 11.0, 7.0]
sol_NaCl = [35.7, 35.9, 36.4, 37.1, 38.0, 39.2]


#regresjon
grad = 2
reg_NaCl = polyfit(T, sol_NaCl, grad)
print(reg_NaCl)

x = linspace(0,250, 10000)
a = reg_NaCl[0]
b = reg_NaCl[1]
c = reg_NaCl[2]
y = a*x**2 + b*x + c


scatter(T,sol_NaCl, color = "blue")
plot(x, y, color = "red")
show()
```

### regresjon av salmiakk
 først endrer jeg listene til arrayer sånn at je gkan bruke r2_score, som regner ut hvor riktig funksjonen er i forhold til punktene. 

```python
from pylab import * 
from sklearn.metrics import r2_score

T = array([0, 20, 40, 60, 80, 100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])


#regresjon
grad = 3
reg_NH3 = polyfit(T, sol_NH3, grad)
#print(reg_NH3)

x = linspace(0,250, 10000)
a = reg_NH3[0]
b = reg_NH3[1]
c = reg_NH3[2]
d = reg_NH3[3]
y = a*x**3 + b*x**2 + c*x + d
y_R2 = a*T**3 + b*T**2 + c*T + d

R2 = r2_score(sol_NH3, y_R2)
print(R2)


scatter(T,sol_NH3, color = "blue")
plot(x, y, color = "red")
grid()
ylim(0)
show()
```
