## 1. derivasjon
### derivasjon definisjon -> den deriverte til en funksjon er et mål på endringa i funksjonsverdien F(x) som følge av en veldig liten endring i variabelen x.
f'(x) = $lim x->0 (f(x+∆x) - f(x))/∆x$


#### promod definisjon
derivasjon er oppfatningen av endringen mellom to punkter. ikke bare matte


```python
from pylab import *

def fder(f, x, delta_x=1 * 10**-7):
    return (f(x + delta_x)-f(x))/delta_x

def f(x):
    return x**2

df = fder(f, int(input()))

print(df)
```

b) En tangent er den momentane veksten i et punkt. det er altså en graf for stigningstallet i et punkt.

$$f'(x) = \lim_{∆a \to 0} \frac{f(a + ∆a) - f(a)}{∆a}$$

Det kan skrives slik fordi vi sier at ∆a går mot null, det vil si at de to punktene går så nærme hverandre det er mulig, omtrent så nærme at de er et punkt. da kan man finne en funksjon for stigningstallet til en graf i helhet.

c) Derivasjon er en oppfattning av endring mellom to punkter.

d) forskjellen mellomdefinisjonen av den deriverte og Newtons kvotient er at Newtons kvotient må h være et tall som er nære null, mens i definisjonen av den deriverte kan vi la $∆a$ gå mot et tall som er så nærme null at vi kan regne med det som null. 

$$f'(x) = \lim_{∆a \to 0} \frac{f(a + ∆a) - f(a)}{∆a}$$

$$f'(x) ≈ \frac{f(x + h) - f(x)}{h}$$
